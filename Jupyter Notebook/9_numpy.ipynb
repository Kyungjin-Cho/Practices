```python
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0

print(a[mask1+mask2])
print(a[mask1*mask2])
```

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
