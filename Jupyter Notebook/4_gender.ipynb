```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('korea_gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in range(0, 101):
            m.append(int(row[i + 3]))
            f.append(int(row[-(i+1)]))
            
f.reverse()
```

```python
f = open('korea_gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:]:
            f.append(int(i))
```

```python
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
f = open('korea_gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title("신도림 지역의 남녀 성별 인구 분포")
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title("신도림 지역의 남녀 성별 인구 분포")
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.style.use('ggplot')
plt.legend()
plt.show()
```

```python
f = open('korea_gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))


plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
